# Problem 2

The formula of an exponential distribution is: $P_Y(y) = \lambda e^{-\lambda y }$
The estimator is obtained as a solution of the maximization problem:
$$\hat{\lambda_n} = l arg max_n l(\lambda; y_1; ... ; y_n)$$
The first order condition for maximum is :
$$\frac{\partial}{\partial \lambda} l(\lambda; y_1; ... ; y_n) = 0 $$
$$ \frac{\partial}{\partial \lambda} l(\lambda; y_1; ... ; y_n) = \frac{\partial}{\partial \lambda}(n*ln(\lambda) - \lambda \sum_{j=1}^n y_j)$$
$$\frac{n}{\lambda} - \sum_{j=1}^{n}y_j = 0 $$
$$\lambda = \frac{n}{\sum_{j=1}^{n}y_j}$$

The estimator of the exponential distribution is $\frac{n}{\sum_{j=1}^{n}y_j}$

```python
## print the gausienne curve

mu = 50
variance = 4
sigma = m.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x,mu,sigma))
plt.title("Distribution of the force needed to break the steel wire ")
plt.xlabel("Force apply to the steel wire (kN)")
plt.ylabel("Probability")
plt.show()
```

```python
wires_dist = stats.norm.pdf(x, mu , sigma)

print("Problem 3 a)")
q = stats.norm.ppf(0.95 , loc= mu, scale= sigma)

interval_value = q - mu
lower_95 = mu - interval_value
upper_95 = mu + interval_value

print("95 % confidence interval", [lower_95, upper_95 ])



plt.plot(x, stats.norm.pdf(x,mu,sigma))
plt.axvline(lower_95, color='r', linestyle=':')
plt.axvline(upper_95, color='r', linestyle=':')
plt.title("Distribution of the force needed to break the steel wire w/ Interval of confidence of 95%")
plt.xlabel("Force apply to the steel wire (kN)")
plt.ylabel("Probability")
plt.show()
```
